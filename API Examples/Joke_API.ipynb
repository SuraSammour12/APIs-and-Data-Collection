take 10 random jokes and tern them into a table

```python
# Send a GET request to the official Joke API
import requests
import json
import pandas as pd

# The API returns 10 random jokes
url = 'http://official-joke-api.appspot.com/jokes/ten'
response = requests.get(url)
```

Data analysis using json.loads()

```python
# Convert the response text to python objects (list if dictionaries)
jokes = json.loads(response.text)
```

```python
# View a sample joke
print(jokes[0])
```

 Convert data to a table (DataFrame) and delete columns

```python
# Convert jokes list into a pandas DataFrame
df_jokes = pd.DataFrame(jokes)

# Drop the 'type' and 'id' columns since they are not needed
df_jokes = df_jokes.drop(columns=['type','id'])

# Show the final DataFrame
df_jokes.head()
```
